```python
#Models
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC, SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#Converting the qualioty label to numeric value
label_quality = LabelEncoder()

df_wines['quality_label'] = label_quality.fit_transform(df_wines['quality_label'])
```

```python
#Selecting the best features
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

#Preparing the data
X = df_wines.drop(['quality', 'wine_category'], axis=1)
y = df_wines['quality_label']

#Apply SelectKBest class to extract top 10 best features
bestfeatures = SelectKBest(score_func=chi2, k=10)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)

#Concat two dataframes for better visualization
featureScores = pd.concat([dfcolumns, dfscores], axis=1)
featureScores.columns = ['Specs', 'Score']
print(featureScores.nlargest(10, 'Score'))
```

```python
c = featureScores.nlargest(10, 'Score')
c['Specs']
```

```python
bf = featureScores.nlargest(10, 'Score')
bfs = bf['Specs']
cols = []
for i in bfs:
  cols.append(i)
```

```python
bfs
```

```python
cols
```

```python
#Selecting the best columns
n_cols = cols[1:]
n_cols
```

```python
#Outliers
for column in n_cols:
    plt.figure(figsize=(8, 6))  # Set the figure size
    sns.boxplot(x=df_wines[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
#Removing the outliers using IQR Rule
df_filtered = df_wines.copy()

# Loop through each column and remove outliers using the IQR rule
for column1 in n_cols:
    Q1 = df_filtered[column1].quantile(0.25)
    Q3 = df_filtered[column1].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 0.4 * IQR
    upper_bound = Q3 + 0.4 * IQR

    # Filter the data to keep only values within the bounds
    df_filtered = df_filtered[(df_filtered[column1] >= lower_bound) & (df_filtered[column1] <= upper_bound)]

#Removed Outliers
for column in n_cols:
    plt.figure(figsize=(8, 6))  # Set the figure size
    sns.boxplot(x=df_filtered[column])
    plt.title(f'Boxplot of {column}')
    plt.xlabel(column)
    plt.show()
```

```python
X = df_filtered[n_cols]
y = df_filtered['quality_label']

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)

models = [LogisticRegression(),
          LinearSVC(),
          SVC(kernel='poly'),
          DecisionTreeClassifier()]
```

```python
model_names = ['LogisticRegression',
               'LinearSVM',
               'Polynomial',
               'DecisionTree']

acc = []
eval_acc = {}

for model in range(len(models)):
  classification_model = models[model]
  classification_model.fit(x_train, y_train)
  pred = classification_model.predict(x_test)
  acc.append(accuracy_score(pred, y_test))

eval_acc = {'Modelling Algorithm': model_names, "Accuracy":acc}
eval_acc
```

```python
acc_table = pd.DataFrame(eval_acc)
acc_table = acc_table.sort_values(by='Accuracy', ascending=[False])
acc_table
```

```python
sns.barplot(y='Modelling Algorithm', x='Accuracy', data=acc_table)
```
